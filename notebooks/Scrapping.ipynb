# Analisis Sentimen Shopee

Data yang diperoleh dari ulasan pengguna aplikasi shopee pada platform google play store. Data tersebut berupa teks review pengguna yang kemudian diperoses analisis sentimen guna mengklasifikasikan opini pengguna ke dalam kategori sentimen

# Scrapping

```python
import pandas as pd
import datetime
import numpy as np

from google_play_scraper import reviews, Sort

app_id = 'com.shopee.id'

def get_reviews(app_id, lang='id', count=20000, sort=Sort.NEWEST):
    try:
        result, continuation_token = reviews(
            app_id,
            lang=lang,
            country='id',
            sort=sort,
            count=count
        )

        return result, continuation_token
    except Exception:
        print("Error:", Exception)
        return None, None

reviews, continuation_token = get_reviews(app_id)

print("jumlah ulasan", len(reviews))
```

```python
# Menyimpan ulasan dalam file CSV
import csv

with open('ulasan_shopee.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])
    for review in reviews:
        writer.writerow([review['content']])
```

```python
data = pd.DataFrame(reviews)
data.shape
data.head()
data.to_csv('ulasan_shopee.csv',index=False)
```
